```python
df["Sexo"].value_counts()
```

- Na coluna sexo, como contém apenas 7 valores faltantes, preferi utillizar a categoria com maior frequência para completar os dois valores faltantes.

```python
# Transformação dos dados de data que estão como objetos
df['PrazoResposta'] = pd.to_datetime(df['PrazoResposta'], format="%Y-%m-%d %H:%M:%S,%f")
df['DataAbertura'] = pd.to_datetime(df['DataAbertura'], format="%Y-%m-%d %H:%M:%S,%f")
df['DataFinalização'] = pd.to_datetime(df['DataFinalização'], format="%Y-%m-%d %H:%M:%S,%f")
df['DataResposta'] = pd.to_datetime(df['DataResposta'], format="%Y-%m-%d %H:%M:%S,%f")
```

```python
# Transformação dos dados de horas que estão como objetos
df["HoraAbertura"] = pd.to_datetime(df["HoraAbertura"], format="%Y-%m-%d %H:%M:%S,%f").dt.strftime('%H:%M:%S')
df["HoraFinalização"] = pd.to_datetime(df["HoraFinalização"], format="%Y-%m-%d %H:%M:%S,%f").dt.strftime('%H:%M:%S')
df["HoraResposta"] = pd.to_datetime(df["HoraResposta"], format="%Y-%m-%d %H:%M:%S,%f").dt.strftime('%H:%M:%S')

```

```python
# Criação de uma nova coluna, se a reclamação foi respondida no prazo
df.loc[df['DataResposta'] <= df["PrazoResposta"], "RespondeuPrazo"] = 'S' 
df.loc[df['DataResposta'] > df["PrazoResposta"], "RespondeuPrazo"] = 'N' 


df["RespondeuPrazo"].value_counts()
```

```python
# Criação de uma nova coluna, se o atendimento foi bom ou ruim com base na nota
df.loc[df['NotaDoConsumidor'] <= 3, "Atendimento"] = 'Ruim' 
df.loc[df['NotaDoConsumidor'] > 3, "Atendimento"] = 'Bom' 


df["Atendimento"].value_counts()
```

```python
# Assim como temos a métrica do tempo de resposta, achei interessante adicionar na análise o tempo para encerramento do processo

df["TempoFinalizar"] = (df["DataFinalização"] - df["DataAbertura"]).dt.days
df["TempoFinalizar"]
```

```python
# Conferir quais dados ainda faltam
df.isnull().sum()
```

```python
df = df.dropna()
```

# 4. Análise Exploratória

Vai ser feito uma análise detalhada das reclamações registradas pelos consumidores, utilizando técnicas estatísticas para identificar padrões e tendências, além de entender a ocorrência de tipos específicos de reclamações.

## 4.1 Variáveis Quantitativas

```python
# Medidas resumo para as variáveis quantitativas
df.describe(include=np.number) # incluindo apenas as quantitativas
```

```python
# Função que cria o gráfico para as variáaveis quantitativaas
def cria_grafico_quant(df,var):

    fig, (ax1,ax2) = plt.subplots(1,2,figsize=(10,5)) # Cria as imagens de fundo

    cor = np.random.rand(3) 
    ax1.hist(df[var],bins=10, color=cor,alpha=0.4) # Cria o histograma
    ax1.title.set_text(f"Histograma de {var}") # Define o titulo
    
    ax2.title.set_text(f"Boxplot de {var}") # Define o titulo do segundo gráfico
    sns.boxplot(data=df[var]) # Cria o Boxplot 
    
# Função que cria a tabela de frequência com as colunas quantitativas agrupadas
def cria_tabela_quanti(df, var):
    copia_df = df.copy() # cria uma copia do dataframe original
    
    if len(df[var].unique()) < 6: # Caso a coluna contenha mais de 6 categorias
        pass
    else :
        copia_df[var] = pd.qcut(df[var], 4) # Corta as categorias em grupos, [0.25, 0.5, 0.75, 1]

    
    df_tab = pd.DataFrame(copia_df[var].value_counts()).sort_values(by=var,ascending=False) # Cria a tabelaa
    total = df_tab.sum() 
    df_tab['Freq_Relativa'] = (df_tab/total).round(decimals=4) 
    df_tab['Freq_Acumulada'] = df_tab['Freq_Relativa'].cumsum().round(decimals=2)
    
    return df_tab
```

```python
df_quanti = df.select_dtypes(include=[np.number]) # cria uma lista com as variáveis quantitativas


for var in df_quanti:
    cria_grafico_quant(df,var)
```

```python
for var in df_quanti:
    print()
    tabela_frequencia = cria_tabela_quanti(df,var)
    print(tabela_frequencia)
```

Com os gráficos e as tabelas das variáveis quantitativas, podemos concluir as informações obtidas que:
- Os anos de 2022 e 2021 representam 66% da base.
- Os dois primeiros e dois últimos meses foram os que mais apresentaram reclamações.
- A média e mediana do tempo para a reposta foi de 4 dias.
- A maioria das reclamações teve uma resposta em até 4 dias.
- O máximo em dias para a resposta foi de 16.
- A média para finalização da reclamação foi de 12 dias e a mediana de 10 dias.
- 75% das reclamações foraam resolvidas em até 20 dias.
- Poucas reclamações se estenderam para mais de 30 dias.
- Referente a avaliação do consumidor, 71% das notas foram distribuidas em 5 (42%) ou em 1 (31%)

# 7. Gráficos auxiliares

## 7.1 Avaliação da Reclamação por Nota

```python
# Definindo as cores
VERMELHO = "#f56369"
AZUL1, AZUL2 = "#B2CFDE", "#086494"
CINZA1, CINZA2, CINZA3 = '#231F20', '#414040', '#76787B'
CINZA4 = '#929497'

# Confirmando os valores
aux = df.groupby(["AvaliaçãoReclamação", "NotaDoConsumidor"])["AvaliaçãoReclamação"].count()
```

```python
X = ["Muito Ruim", "Ruim", "Neutro", "Bom", "Muito Bom"] # Valores referentes as notas
Y = aux["Não Resolvida"].values  # notas das avaliações não resolvidas
Z = aux["Resolvida"].values  # notas das avaliações resolvidas


# cria uma nova figura
plt.figure(figsize=(7.45, 4.9), dpi=110)

plt.subplots_adjust(left=0, right=0.735, top=0.665, bottom=0.17)


def subcategorybar(X,  # Categorias
                   vals,  # Valores numericos ou subcategorias
                   base_colors,  # Cor base das categorias
                   special_colors,  # Cores para categorias especiais
                   special_bars, # Indice de qual categoria é especial
                   color_annotate, #Cor para rotulo
                   width=0.9):  # Tamanho da barra
    n = len(vals)
    _X = np.arange(len(X))
    
    for i in range(n):
        # Cria as barras com os valores númericos
        bars = plt.bar(_X - width / 2. + i / float(n) * width,
                       vals[i],
                       width=width / float(n),
                       align="edge",  # Alinha as barras a esquerda
                       color=base_colors[i])  # usa a cor para barra

        # Rotulo dos dados
        for bi, b in enumerate(bars):
            plt.text(b.get_x() + 0.05, b.get_height() - 3.5,
                     str(vals[i][bi]),
                     fontsize=12,
                     color=color_annotate)
            
        # Barras especiais
        if special_bars[i] != None:
            for special_bar_i in special_bars[i]:
                bars[special_bar_i].set_color(special_colors[i])

    
    # Seta as categorias em seus locais e seus respectivos valores
    plt.xticks(_X, X)

# Chama a função pra criar o gráfico
subcategorybar(X,  # Resposta
               [Y, Z],  # Valores das Avaliações não resolvidas e resolvidas respectivamente
               [CINZA3, CINZA3],  # cores para as barras normais
               [VERMELHO, AZUL2],  # Cores para as barras especiais
               [[0], [3, 4]],  # Destaca as barras pedidas
                    CINZA2) # cor para o rótulo da categoria

# remove as bordas do gráfico
for spine in plt.gca().spines.values():
    spine.set_visible(False)

# Desabilita as linhas e valoers das categorias de y e a linha de x
plt.tick_params(bottom=False, left=False, labelleft=False)



# Cria o título do gráfico
plt.text(-0.65, 70000, 'Nota do Atendimento de acordo com a Resolução da Reclamação', fontsize=15, 
         color=CINZA1);

# Legenda Não Resolvidas
plt.text(0, 60000,
         '$\\bf{NÃO}$ Resolvida',  # mathtext \\bf para negrito
         fontsize=14,
         linespacing=1.3,  # espaço para a fonte
         color=VERMELHO);

# Legenda Resolvidas
plt.text(4.6, 10,
         '$\\bf{RESOLVIDA}$\n',  # mathtext \\bf para negrito
         fontsize=14,
         linespacing=1.45,  
         color=AZUL2);

plt.savefig('./gráficos/ReclamaçõesAtendimento.png')
```
